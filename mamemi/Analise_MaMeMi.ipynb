# Analisando os resultados do detector MaMeMi

```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

```python
import json
```

```python
with open('data.json','r') as fp:
    data = json.load(fp)
```

```python
patient = data['111']
```

```python
# ecg_file = 'ecg.npy'

# x = np.load(ecg_file) #TODO trocar para json
# N = x.shape[0]

x = np.array(patient['MLII'][17000:20000])
N = x.shape[0]
```

```python
#contagem de percentual de execucao
ip = 0
lg = 0

# R detector variables
sigma = 2
delta = 2
beta = 15
atv = 0

maxi = np.zeros(N+beta)
mini = np.zeros(N+beta)
h = np.zeros(N+beta)
a = np.zeros(N+beta)
g = np.zeros(N+beta)
n = np.zeros(N+beta)
r = np.zeros(N+beta)
v = np.zeros(N+beta)
w = np.zeros(N+beta)

maxi[0] = x[0]
mini[0] = x[0]
h[0] = x[0] - (maxi[0]+mini[0])/2

for i in range(1,N):
    # calculando maxi e mini
    if x[i] > maxi[i-1]:
        maxi[i] = maxi[i-1] + sigma*delta
    else:
        maxi[i] = maxi[i-1] - delta

    if x[i] < mini[i-1]:
        mini[i] = mini[i-1] - sigma*delta
    else:
        mini[i] = mini[i-1] + delta

    # calculando o h
    h[i] = x[i] - (maxi[i] + mini[i])/2

    # calculando a
    a[i] = maxi[i] - mini[i]

    if a[i] <= abs(h[i]):
        n[i] = np.sign(h[i])*(abs(h[i])-a[i])
    else:
        n[i] = 0

    if i > beta:
        if n[i]>0 and n[i-beta]<n[i] and n[i]>n[i+beta]:
            g[i] = n[i] - max(n[i-beta], n[i+beta])
        elif n[i]<0 and n[i-beta]>n[i] and n[i]<n[i+beta]:
            g[i] = n[i] + min(n[i-beta], n[i+beta])
        else:
            g[i] = 0

    if g[i] > g[i-1] and g[i] > g[i+1]:
        r[i] = g[i]
    else:
        r[i] = 0

    if g[i] < g[i-1] and g[i] < g[i+1]:
        v[i] = g[i]
    else:
        v[i] = 0

    if r[i] > 0:
        w[i] = r[i]
    elif v[i] < 0:
        w[i] = -v[i]

    if i > ip:
        print("executando...", lg)
        lg += 10
        ip += (N-1)/10
```
